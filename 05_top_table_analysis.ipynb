# TOP TABLES - VISUALIZATION AND INSIGHTS

```python
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

engine = create_engine("mysql+mysqlconnector://root:1234@localhost/phonepe")

query = "SELECT state, SUM(registered_users) AS total_users FROM top_user GROUP BY state ORDER BY total_users DESC;"
df_users = pd.read_sql(query, engine)
```

## Top States by Registered Users

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_users, y='state', x='total_users', color='orange')  
plt.title("Top States by Registered Users")
plt.xlabel("Total Users")
plt.ylabel("State")
plt.tight_layout()
plt.savefig("C:/Users/sharan/Documents/phonepe_transaction_insights/images/top_states_registered_users.png")
plt.show()
```

## INSIGHTS:
- Maharashtra leads in registered users, likely due to urban centers like Mumbai and Pune with strong digital infrastructure.
- Uttar Pradesh, with its huge population, shows impressive digital reach, signaling successful penetration in tier-2 and tier-3 cities
