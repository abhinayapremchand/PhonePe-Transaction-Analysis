```python
# ------------------- Libraries -------------------
import pandas as pd
import mysql.connector
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# ------------------- DB Connection -------------------
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="1234",
    database="phonepe_insights"
)

# ------------------- Load Tables -------------------
tables = [
    "aggregated_user",
    "aggregated_transaction",
    "aggregated_insurance",
    "map_user",
    "map_transaction",
    "map_insurance",
    "top_user",
    "top_transaction",
    "top_insurance"
]

dataframes = {}
for table in tables:
    df = pd.read_sql(f"SELECT * FROM {table};", conn)
    # drop completely null rows if any
    df = df.dropna(how="all")
    dataframes[table] = df
    print(f"\n--- {table.upper()} ---")
    print(df.head())
    print(df.info())

print("\nData loaded and cleaned")
```

```python
import os

# folder to save plots
save_path = r"C:\Users\sharan\Documents\phonepe_insights\graphs"
os.makedirs(save_path, exist_ok=True)

# ---------------- Aggregated Transactions ----------------
agg_trans = dataframes["aggregated_transaction"]

# yearly transaction trends
yearly_trans = agg_trans.groupby("year")[["count", "amount"]].sum().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(data=yearly_trans, x="year", y="count", marker="o", label="Transactions")
sns.lineplot(data=yearly_trans, x="year", y="amount", marker="o", label="Amount (₹)")
plt.title("Yearly Transaction Trends (All India)")
plt.ylabel("Count / Amount")
plt.legend()
plt.savefig(os.path.join(save_path, "yearly_transaction_trends.png"))
plt.show()

# top 10 states by total transaction amount
top_states_trans = agg_trans.groupby("state")["amount"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_states_trans.values, y=top_states_trans.index, hue=top_states_trans.index,
            dodge=False, palette="viridis", legend=False)
plt.title("Top 10 States by Transaction Amount")
plt.xlabel("Total Amount (₹)")
plt.savefig(os.path.join(save_path, "top_states_transactions.png"))
plt.show()


# ---------------- Aggregated Users ----------------
agg_user = dataframes["aggregated_user"]

# yearly registered users trend
yearly_users = agg_user.groupby("year")[["registered_users", "app_opens"]].sum().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(data=yearly_users, x="year", y="registered_users", marker="o", label="Registered Users")
sns.lineplot(data=yearly_users, x="year", y="app_opens", marker="o", label="App Opens")
plt.title("Yearly Growth of Users & App Opens (All India)")
plt.ylabel("Count")
plt.legend()
plt.savefig(os.path.join(save_path, "yearly_users_growth.png"))
plt.show()

# top 10 states by registered users
top_states_users = agg_user.groupby("state")["registered_users"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_states_users.values, y=top_states_users.index, hue=top_states_users.index,
            dodge=False, palette="magma", legend=False)
plt.title("Top 10 States by Registered Users")
plt.xlabel("Total Registered Users")
plt.savefig(os.path.join(save_path, "top_states_users.png"))
plt.show()


# ---------------- Aggregated Insurance ----------------
agg_insurance = dataframes["aggregated_insurance"]

# yearly insurance trend
yearly_insurance = agg_insurance.groupby("year")[["count", "amount"]].sum().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(data=yearly_insurance, x="year", y="count", marker="o", label="Insurance Count")
sns.lineplot(data=yearly_insurance, x="year", y="amount", marker="o", label="Insurance Amount (₹)")
plt.title("Yearly Insurance Trends (All India)")
plt.ylabel("Count / Amount")
plt.legend()
plt.savefig(os.path.join(save_path, "yearly_insurance_trends.png"))
plt.show()

# top states in insurance by amount
top_states_ins = agg_insurance.groupby("state")["amount"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_states_ins.values, y=top_states_ins.index, hue=top_states_ins.index,
            dodge=False, palette="coolwarm", legend=False)
plt.title("Top 10 States by Insurance Amount")
plt.xlabel("Total Insurance Amount (₹)")
plt.savefig(os.path.join(save_path, "top_states_insurance.png"))
plt.show()
```

### INSIGHTS
-  Yearly Transaction Trends – Transaction amounts have steadily increased, peaking in 2024, while transaction counts remained flat, suggesting growth is driven by higher value per transaction.
- Top 10 States by Transaction Amount – Telangana, Karnataka, and Maharashtra lead in transaction volumes, highlighting them as key markets.
- Yearly Growth of Users & App Opens – Registered users remained stable, but app engagement (app opens) has grown significantly, with 2024 showing the highest engagement.
- Top 10 States by Registered Users – Maharashtra and Uttar Pradesh dominate user registrations, reflecting dense adoption.
- Yearly Insurance Trends – Insurance amounts have risen year after year, but the number of policies remains flat, showing higher premium values rather than more customers.
- Top 10 States by Insurance Amount – Karnataka and Maharashtra are the leaders in insurance adoption, indicating financial awareness in these states.

```python

# --- Folder to save graphs ---
graph_folder = r"C:\Users\sharan\Documents\phonepe_insights\graphs"
os.makedirs(graph_folder, exist_ok=True)

# --- Load data from MySQL ---
map_transaction = pd.read_sql("SELECT * FROM map_transaction;", conn)
map_user = pd.read_sql("SELECT * FROM map_user;", conn)
map_insurance = pd.read_sql("SELECT * FROM map_insurance;", conn)
agg_transaction = pd.read_sql("SELECT * FROM aggregated_transaction;", conn)
agg_insurance = pd.read_sql("SELECT * FROM aggregated_insurance;", conn)

#top 10 states by total amoount ( Transactions + Insurance ) and Registered Users
top10_states_users = map_user.groupby("state")["registered_users"].sum().nlargest(10).index
top10_states_trans = agg_transaction.groupby("state")["amount"].sum().nlargest(10).index
top10_states_ins = agg_insurance.groupby("state")["amount"].sum().nlargest(10).index

# =====================
# 1. Yearly Users per State
# =====================
yearly_users = map_user.groupby(["year", "state"])["registered_users"].sum().reset_index()
yearly_users = yearly_users[yearly_users["state"].isin(top10_states_users)]

plt.figure(figsize=(14,6))
sns.barplot(data=yearly_users, x="year", y="registered_users", hue="state", dodge=True)
plt.title("Yearly Registered Users (Top 10 States)")
plt.xlabel("Year")
plt.ylabel("Registered Users")
plt.legend(title="State", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.savefig(os.path.join(graph_folder, "yearly_users_top10.png"))
plt.show()
plt.close()
# =====================
# 2. Yearly Transactions Amount per State (Top 10 States)
# =====================
yearly_trans = agg_transaction.groupby(["year", "state"])["amount"].sum().reset_index()
yearly_trans = yearly_trans[yearly_trans["state"].isin(top10_states_trans)]

plt.figure(figsize=(14,6))
sns.barplot(data=yearly_trans, x="year", y="amount", hue="state", dodge=True)
plt.title("Yearly Transactions Amount (Top 10 States)")
plt.xlabel("Year")
plt.ylabel("Total Amount (INR)")
plt.legend(title="State", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.savefig(os.path.join(graph_folder, "yearly_transactions_top10.png"))
plt.show()
plt.close()

# =====================
# 3. Yearly Insurance Amount per State (2020+, Top 10 States)
# =====================
yearly_ins = agg_insurance[agg_insurance["year"] >= 2020].groupby(["year", "state"])["amount"].sum().reset_index()
yearly_ins = yearly_ins[yearly_ins["state"].isin(top10_states_ins)]

plt.figure(figsize=(14,6))
sns.barplot(data=yearly_ins, x="year", y="amount", hue="state", dodge=True)
plt.title("Yearly Insurance Amount (2020+, Top 10 States)")
plt.xlabel("Year")
plt.ylabel("Total Amount (INR)")
plt.legend(title="State", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.savefig(os.path.join(graph_folder, "yearly_insurance_top10.png"))
plt.show()
plt.close()

# =====================
# 4. Top 10 Districts by Transactions
# =====================
top_districts_trans = map_transaction.groupby("district")["amount"].sum().nlargest(10).reset_index()
plt.figure(figsize=(10,6))
sns.barplot(data=top_districts_trans, x="amount", y="district", palette="cubehelix")
plt.title("Top 10 Districts by Transactions Amount")
plt.tight_layout()
plt.savefig(os.path.join(graph_folder, "top10_dist_transactions.png"), bbox_inches="tight")
plt.show()
plt.close()

# =====================
# 5. Top 10 Districts by Registered Users
# =====================
top_districts_users = map_user.groupby("district")["registered_users"].sum().nlargest(10).reset_index()
plt.figure(figsize=(10,6))
sns.barplot(data=top_districts_users, x="registered_users", y="district", palette="cubehelix")
plt.title("Top 10 Districts by Registered Users")
plt.tight_layout()
plt.savefig(os.path.join(graph_folder, "top10_dist_users.png"), bbox_inches="tight")
plt.show()
plt.close()


# =====================
# 6. Top 10 States by Insurance Amount (from 2020 only)
# =====================
top_states_ins = agg_insurance[agg_insurance["year"] >= 2020].groupby("state")["amount"].sum().nlargest(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_states_ins.values, y=top_states_ins.index, hue=top_states_ins.index, palette="cubehelix", legend=False)
plt.title("Top 10 States by Insurance Amount (2020+)")
plt.tight_layout()
plt.savefig(os.path.join(graph_folder, "top10_states_insurance.png"))
plt.show()
plt.close()
```

## CONCLUSION:
The EDA highlights consistent digital adoption growth across India, with **2024 marking the peak year**.  
Transaction amounts are rising faster than transaction counts, suggesting that **users are spending more per transaction**.  

At the state level, **Karnataka, Maharashtra, Telangana, and Uttar Pradesh** dominate in transactions, user registrations, and insurance adoption.  
At the district level, **Bengaluru Urban, Pune, Hyderabad, and Thane** stand out as key hubs of digital engagement.  

Insurance adoption is **concentrated in financially aware regions**, with **Karnataka overtaking Maharashtra post-2022** as the leading state.
